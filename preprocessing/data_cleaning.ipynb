```python
from sklearn.preprocessing import OneHotEncoder

def one_hot_encode_and_add(df, column):
    one_hot_encoder = OneHotEncoder(sparse_output=False)
    one_hot_encoded = one_hot_encoder.fit_transform(df[[column]])
    encoded_columns = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out([column]))
    encoded_columns.index = df.index
    df = pd.concat([df, encoded_columns], axis=1)
    df = df.drop(columns=[column])
    return df

columns_to_encode = ['gender', 'marital_status','Have you made online purchases during crisis time?']

for col in columns_to_encode:
    df = one_hot_encode_and_add(df, col)
```

```python
from collections import Counter
import re
import pandas as pd
from fuzzywuzzy import process        

def smart_encode_multilabel_column(
        df: pd.DataFrame,
        column_name: str,
        canonical_keywords: list[str],
        *,
        threshold: int = 65,             # fuzzy-match cut-off
        min_count: int = 5,              # own column only if it appears ≥ this many times
        prefix: str | None = None        # NEW → column-name prefix (e.g. “general”)
) -> pd.DataFrame:

    # ------------------------------------------------------------------
    # 1.  convert obvious blanks → literal string 'unknown'
    # ------------------------------------------------------------------
    invalid = {'', 'n/a', 'na', 'none', 'no', 'nan',
               'didn’t used', 'didnt used', 'not applicable', "Didn't order any pharmacy items online",'No experience in this flatform'}
    col = (df[column_name]
             .fillna('unknown')
             .astype(str))

    col = col.apply(lambda x: 'unknown'
                    if x.strip().lower() in invalid else x)

    # ------------------------------------------------------------------
    # 2.  helpers for normalising + fuzzy matching
    # ------------------------------------------------------------------
    def _norm(txt: str) -> str:
        txt = re.sub(r'\.com|\.lk|/', '', txt, flags=re.I)   # drop domains/slashes
        txt = re.sub(r'[^A-Za-z0-9 ]+', ' ', txt)           # keep letters + digits
        return txt.lower().strip()

    def _match_cell(cell: str) -> list[str]:
        """Return list of matched keywords for one cell."""
        if cell == 'unknown':
            return ['unknown']

        tokens = [_norm(t) for t in cell.split(',') if t.strip()]
        matches = []
        for tok in tokens:
            match, score = process.extractOne(tok, canonical_keywords)
            matches.append(match if score >= threshold else '__tmp_other__')
        return matches

    # ------------------------------------------------------------------
    # 3.  fuzzy-match every row  →  list of keywords
    # ------------------------------------------------------------------
    tmp_lists = col.apply(_match_cell)

    # ------------------------------------------------------------------
    # 4.  find which keywords meet the frequency threshold
    # ------------------------------------------------------------------
    flat = [w for lst in tmp_lists for w in lst
            if w not in ('unknown', '__tmp_other__')]
    counts = Counter(flat)
    top_kw = {kw for kw, c in counts.items() if c >= min_count}

    # ------------------------------------------------------------------
    # 5.  collapse rare matches + tmp_other into 'other'
    # ------------------------------------------------------------------
    def _finalise(lst: list[str]) -> list[str]:
        out = []
        for w in lst:
            if w == 'unknown':
                out.append('unknown')
            elif w in top_kw:
                out.append(w)
            else:
                out.append('other')
        # remove duplicates while preserving order
        seen = set()
        return [x for x in out if not (x in seen or seen.add(x))]

    final_lists = tmp_lists.apply(_finalise)

    # keep a column with the clean lists (optional but handy)
    df[f'{column_name}_clean'] = final_lists

    # ------------------------------------------------------------------
    # 6.  multi-hot encode
    # ------------------------------------------------------------------
    # decide the prefix once
    if prefix is None:
        base = f'use_{column_name}_'       # old behaviour
    else:
        base = f'{prefix}_'                # e.g. "general_"

    for kw in sorted(top_kw | {'other', 'unknown'}):
        df[f'{base}{kw}'] = final_lists.apply(lambda lst, k=kw: int(k in lst))

    df.drop(columns=[column_name], inplace=True)
    
    return df
```

```python
def standardize_and_encode_likert(df, columns):
 
    # Normalize spelling/capitalization
    normalization_map = {
        'Strongly agree': '5',
        'Strongly Agree': '5',
        'Agree': '4',
        'Neutral': '3',
        'Disagree': '2',
        'Strongly disagree': '1',
        'Srongly Disagree': '1',  
    }

    for col in columns:
        df[col] = df[col].replace(normalization_map)            

    return df
```
