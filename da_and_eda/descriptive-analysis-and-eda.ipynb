```python
import pandas as pd
import matplotlib.pyplot as plt


# Columns to create pie charts for
columns = [
    "EaseofNavigation1", "EaseofNavigation2", "LearnCurve1", "LearnCurve2",
    "ClarityofInstructions1", "ClarityofInstructions2", "ResponseTime1", "ResponseTime2",
    "ErrorHandling1", "ErrorHandling2", "ProductAvailability1", "ProductAvailability2",
    "Convenience1", "Convenience2", "Cost-effectiveness1", "Cost-effectiveness2",
    "InformationAccessibility1", "InformationAccessibility2", "Personalization1", "Personalization2",
    "PrivacyProtection1", "PrivacyProtection2", "PaymentSecurity1", "PaymentSecurity2",
    "TransparentPolicies1", "TransparentPolicies2", "WordofMouthandRecommendations1", "WordofMouthandRecommendations2",
    "SocialMediaPosts/Influence1", "SocialMediaPosts/Influence2", "OnlineReviewsandRatings1", "OnlineReviewsandRatings2",
    "SocialProof1", "SocialProof2", "NormativePressure1", "NormativePressure2", "InformationSharing1", "InformationSharing2",
    "Attitude1", "Attitude2", "PerceivedRisk1", "PerceivedRisk2", "IntentiontoPurchaseOnline1", "IntentiontoPurchaseOnline2",
    "IntentiontoPurchaseOnline3", "IntentiontoPurchaseOnline4"
]

import matplotlib.pyplot as plt

for col in columns:
    counts = df[col].value_counts().sort_index()
    labels = [f"{int(val)} ({cnt})" for val, cnt in counts.items()]
    plt.figure(figsize=(5, 5))
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
categorical_columns = ['age', 'education', 'used_before', 'profession']

# Plot bar charts for each categorical column
for col in categorical_columns:
    counts = df[col].value_counts().sort_index()
    plt.figure(figsize=(8, 5))
    ax = sns.barplot(x=counts.index.astype(str), y=counts.values)

    # Add count labels on each bar
    for i, value in enumerate(counts.values):
        ax.text(i, value + 0.5, str(value), ha='center', va='bottom')

    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Count how many rows have a value of 1 for each gender column
gender_counts = {
    'Female': df['gender_Female'].sum(),
    'Male': df['gender_Male'].sum()
}

# Plot the gender bar graph
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=list(gender_counts.keys()), y=list(gender_counts.values()))

# Add count labels
for i, value in enumerate(gender_counts.values()):
    ax.text(i, value + 0.5, str(int(value)), ha='center', va='bottom')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Count how many rows have a value of 1 for each crisis purchase response column
purchase_during_crisis_counts = {
    'No': df['Have you made online purchases during crisis time?_No'].sum(),
    'Yes': df['Have you made online purchases during crisis time?_Yes'].sum()
}

# Plot the bar chart
plt.figure(figsize=(6, 4))
ax = sns.barplot(x=list(purchase_during_crisis_counts.keys()), y=list(purchase_during_crisis_counts.values()))

# Add count labels
for i, value in enumerate(purchase_during_crisis_counts.values()):
    ax.text(i, value + 0.5, str(int(value)), ha='center', va='bottom')

plt.title("Online Purchases During Crisis Time")
plt.xlabel("Response")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Columns representing general platform usage
platform_columns = [
    'general_aliexpress', 'general_daraz', 'general_ebay', 'general_ikman',
    'general_kapruka', 'general_other', 'general_takas', 'general_unknown',
    'general_wishque', 'general_wow'
]

# Sum values across each column to count "used" platforms
platform_counts = df[platform_columns].sum().sort_values(ascending=False)

# Plot the platform usage bar chart
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=platform_counts.index.str.replace('general_', ''), y=platform_counts.values)

# Add count labels
for i, value in enumerate(platform_counts.values):
    ax.text(i, value + 0.5, str(int(value)), ha='center', va='bottom')

plt.title("Usage of Online Purchasing Platforms")
plt.xlabel("Platform")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
